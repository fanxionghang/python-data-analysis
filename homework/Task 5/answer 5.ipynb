```python
import pandas as pd
import os 
os.listdir()
```

```python
titanic=pd.read_csv('Titanic.csv',skiprows=3)
titanic2=pd.read_csv('titanic2.csv')
titanic.head();titanic2.tail()
list(titanic);list(titanic2.columns.values)
titanic.shape;titanic2.shape
```

```python
titanic_combine=pd.concat([titanic,titanic2],ignore_index=True)
```

```python
titanic_indexed=titanic_combine.set_index(['Sex','Pclass'])
```

```python

titanic_indexed.head()
```

```python
titanic_indexed_order=titanic_indexed.swaplevel('Sex','Pclass')
titanic_indexed_order.head()
```

```python
titanic_indexed_order_sort=titanic_indexed_order.sort_index(level=0)
```
